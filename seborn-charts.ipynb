### Libs Import

```python
#LIBS
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Database Import

```python
df_videos = pd.read_excel('Database/videosYT.xlsx', 'videos')
display(df_videos)
```

### Scatter Plot

```python
scatter_plot = sns.scatterplot(data = df_videos, x = 'Nº de Views', y = 'Nº de Likes', hue = 'Categoria', style = 'Responsável')
plt.title('Análise de Nº de Views x Nº de Likes por Categoria e Responsável')
plt.show(scatter_plot)
```

### Relational Plot

```python
relational_plot = sns.relplot(data = df_videos, x = 'Nº de Views', y = 'Nº de Likes', hue = 'Categoria', col = 'Responsável')
relational_plot.set_titles('Responsável: {col_name}')
plt.show(relational_plot)
```

### Line Plot

```python
df_inscritos = pd.read_excel("Database/videosYT.xlsx", "Inscritos")
line_plot = sns.lineplot(data = df_inscritos, x = 'Mês/Ano', y = 'Inscritos', color = 'g')
```

### Histograms

```python
hist_plot = sns.displot(data = df_videos, x = 'Nº de Views', hue = 'Responsável', col = 'Responsável', rug = True)
hist_plot.set_titles('Responsável: {col_name}')
plt.show(hist_plot)
```

```python
#kind = type of histogram (kde = Density)
hist_plot = sns.displot(data = df_videos, x = 'Nº de Views', hue = 'Responsável', col = 'Responsável', rug = True, kind = 'kde')
hist_plot.set_titles('Responsável: {col_name}')
plt.show(hist_plot)
```

```python
#kind = type of histogram (ecdf = percet distribuition)
hist_plot = sns.displot(data = df_videos, x = 'Nº de Views', hue = 'Responsável', col = 'Responsável', rug = True, kind = 'ecdf')
hist_plot.set_titles('Responsável: {col_name}')
plt.show(hist_plot)
```
