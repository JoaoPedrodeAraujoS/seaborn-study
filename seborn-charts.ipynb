### Libs Import

```python
#LIBS
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Database Import

```python
df_videos = pd.read_excel('Database/videosYT.xlsx', 'videos')
display(df_videos)
```

### Scatter Plot

```python
fig = sns.scatterplot(data = df_videos, x = 'Nº de Views', y = 'Nº de Likes', hue = 'Categoria', style = 'Responsável')
plt.title('Análise de Nº de Views x Nº de Likes por Categoria e Responsável')
plt.show(fig)
```
