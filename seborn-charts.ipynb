### Libs Import

```python
#LIBS
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Database Import

```python
df_videos = pd.read_excel('Database/videosYT.xlsx', 'videos')
display(df_videos)
```

### Scatter Plot

```python
scatter_plot = sns.scatterplot(data = df_videos, x = 'Nº de Views', y = 'Nº de Likes', hue = 'Categoria', style = 'Responsável')
plt.title('Análise de Nº de Views x Nº de Likes por Categoria e Responsável')
plt.show(scatter_plot)
```

### Relational Plot

```python
relational_plot = sns.relplot(data = df_videos, x = 'Nº de Views', y = 'Nº de Likes', hue = 'Categoria', col = 'Responsável')
relational_plot.set_titles('Responsável: {col_name}')
plt.show(relational_plot)
```
